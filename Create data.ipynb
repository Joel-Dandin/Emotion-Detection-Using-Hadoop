```python
import pandas as pd

df = pd.read_json("Arts_Crafts_and_Sewing.json", lines=True)
```

```python
df
```

```python
df = df.drop(['vote','verified','reviewTime','reviewerID','style','reviewerName','unixReviewTime','image'], inplace=True, axis=1)
```

```python
df
```

```python
null_data = df[df.isnull().any(axis=1)]
```

```python
null_data
```

```python
df = df.dropna()
```

```python
df
```

```python
ans = [y for x, y in df.groupby('asin')]
```

```python
ans[0]['reviewText']
```

```python
print(len(ans))
```

```python
for i in range (len(ans)):
  print(str(i) + " == " + str(len(ans[i])))
  if(len(ans[i])<1000):
    print(str(i) + " = " + str(len(ans[i])))
    ans.pop(i)
    i = i-3
```

```python
print(len(ans))
```

```python
for i in range (len(ans)):
  ans[i] = ans[i].reset_index()
  print(ans[i].head())
```

```python
print(len(ans))
```

```python
!mkdir /Review
!mkdir /Overall
!mkdir /Summary
```

```python
for i in range (len(ans)):
  single_ans =  ans[i]
  tempans = ans[i].to_csv(columns=['reviewText'], header=None, index=None)
  #print(tempans)
  filename = "/Review/" + single_ans['asin'][0] + ".txt"
  f1 = open(filename,"a+")
  f1.write(tempans)
  
```

```python
for i in range (len(ans)):
  single_ans1 =  ans[i]
  tempans1 = ans[i].to_csv(columns=['overall'], header=None, index=None)
  #print(tempans)
  filename1 = "/Overall/" + single_ans1['asin'][0] + ".txt"
  f1 = open(filename1,"a+")
  f1.write(tempans1)
```

```python
for i in range (len(ans)):
  single_ans2 =  ans[i]
  tempans2 = ans[i].to_csv(columns=['summary'], header=None, index=None)
  #print(tempans)
  filename2 = "/Summary/" + single_ans2['asin'][0] + ".txt"
  f1 = open(filename2,"a+")
  f1.write(tempans2)
```
